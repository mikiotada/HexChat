# Data Simulation

This notebook try to randomly generate the data for algorithm training purpose. It doesn't have to be very consistent nor logical, they provide a data structure / format for us to carry on the training

```python
import pandas as pd
import numpy as np
from faker import Faker
from data.fake_data import *
from datetime import date
```

```python
sample_size = 120

random_state = 42

fake = Faker(['zh_CN', 'en_US', 'ja_JP', 'en_IN', 'de_DE', 'en_GB'])

Faker.seed(random_state)
np.random.seed(random_state)

# # # # # # # # # # # # #
# simulation portion
# # # # # # # # # # # # #

names = [fake.name() for _ in range(sample_size)]
sim_subject = np.random.choice(subjects, sample_size)
sim_school = [np.random.choice(schools, np.random.randint(1, 3)) for _ in range(sample_size)]
sim_gradyr = [np.random.choice(list(range(2010, 2021)), len(sim_school[i]), replace=False) for i in range(sample_size)]
sim_zip = np.random.choice(zips, sample_size)
sim_industries = np.random.choice(industries, sample_size)
sim_titles = np.random.choice(titles, sample_size)
sim_countries = np.random.choice(countries, sample_size)
sim_languages = [np.random.choice(languages, np.random.randint(1, 4), replace=False) for _ in range(sample_size)]

################
# Add on 22 Apr, to better fit the latest product goal, include two more columns for simulation
################

# to simulate the situation that some ppl are connected already, avoid seeing them again
sim_connected = [np.random.choice(range(0, sample_size), i) for i in np.random.choice(range(20), sample_size)]

# priority tags, to simulate the situation that people select the tags explicitly
sim_tags = [['tag_' + str(i) for i in np.random.choice(range(0, 20), i)] for i in np.random.choice(range(4), sample_size)]
```

```python
sim_data = pd.DataFrame(np.vstack([
    names, 
    sim_school,
    sim_subject, 
    sim_gradyr, 
    sim_zip, 
    sim_industries, 
    sim_titles,
    sim_countries,
    sim_languages,
    sim_connected,
    sim_tags
])).T

sim_data.columns = ['name', 'school', 'subject', 'grad_yr', 'zip', 'industry', 'title', 'country', 'language', 'connected', 'priority_tags']

sim_data.reset_index()
```

```python
file_name = 'data/sim_data_' + str(sim_data.shape[0]) + '_' + date.today().strftime("%d%b%Y") + '.csv'

sim_data.to_csv(file_name, index=False)
```
