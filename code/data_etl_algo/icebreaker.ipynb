```python
def icebreaker_similarly(data, ii_input, group_set):
    tmp = combined_inverted_index(data.loc[data['index'].isin(list(group_set))], ii_input, uid_field_name="index")
    
    result = list()
    for col in tmp.keys():
        for key in tmp[col].keys():
            if len(tmp[col][key]) > 1:
                result.append([col, key, len(tmp[col][key])])

    return sorted(result, key=itemgetter(2), reverse=True)
```

```python
def icebreaker_sentense(icebreaker_list):
    for item in icebreaker_list:
        if item[0] == 'priority_tags':
            print(f'{item[2]} of you selected {item[1]}')
        elif item[0] == 'language':
            print(f'{item[2]} of you know {item[1]}')
        elif item[0] == 'zip':
            print(f'{item[2]} of you live around {item[1]}')
        elif item[0] == 'industry':
            print(f'{item[2]} of you work in {item[1]} industry')
        elif item[0] == 'school':
            print(f'{item[2]} of you studied in {item[1]}')
        elif item[0] == 'country':
            print(f'{item[2]} of you have stayed in {item[1]}')
        elif item[0] == 'subject':
            print(f'{item[2]} of you studied {item[1]}')
        elif item[0] == 'title':
            print(f'{item[2]} of you are working as {item[1]}')
        
```
